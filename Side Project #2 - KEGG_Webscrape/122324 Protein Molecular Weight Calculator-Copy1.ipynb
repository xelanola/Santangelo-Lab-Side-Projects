```python
#install packages
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import csv
import re

#make a list and dataframe of all the names of the TK genes

TK_List_1000 =  range(1000,2307)
TK_List_100 =  range(100,1000)
TK_List_10 =  range(10, 100)
TK_List_1 = range(1,10)


TK_1000 = [str(x) for x in TK_List_1000]
TK_100 = [str(x) for x in TK_List_100]
TK_10 = [str(x) for x in TK_List_10]
TK_1 = [str(x) for x in TK_List_1]


TK_0001 = ["000" + x for x in TK_1]
TK_0010 = ["00" + x for x in TK_10]
TK_0100 = ["0" + x for x in TK_100]

TK_genes = TK_0001 + TK_0010 + TK_0100 + TK_1000


TK_GENES = ["TK" + x for x in TK_genes]

Tko_Protein_MW = pd.DataFrame(TK_GENES)
Tko_Protein_MW.rename(columns={0 :'Gene'}, inplace = True)

                      
#Amino Acid Dictionary of Molecular Weights 

amino_acids = {"A": 89, "R": 174, "N": 132, "D": 133, "C": 121, "Q": 146, "E": 147,
               "G": 75, "H": 155, "I": 131, "L": 131, "K": 146, "M": 149, "F": 165,
               "P": 115, "S": 105, "T": 119, "W": 204, "Y": 181, "V": 117}

```

```python
#Web Scrape All Tko Genes for molecular weights and protein functions

protein_value = []
protein_info = []
protein_sequence = []

x=0
for i in TK_GENES:  
    protein_MW = 0
    protein_AAs = []
    protein_seq = []
    URL = "https://www.genome.jp/entry/-f+-n+a+tko:" + i
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    AA = soup.find("pre")
    TKAA=str(AA)
    Almost = TKAA.replace('</pre>', '')
    Almost_Info = (Almost.rsplit('(A)')[0])
    Info = (Almost_Info.rsplit(';tko:')[1])
    protein_info.append(Info)
    Almost_seq = Almost.rsplit('(A)')[1]
    protein_seq.append(Almost_seq.replace("\n", ""))
    protein_sequence.append(protein_seq)
    for j in protein_seq:
        for k in j:
            if k in amino_acids:
                protein_AAs.append(amino_acids[k])
    for l in protein_AAs:
        protein_MW+=l-18
    protein_value.append(int(protein_MW /1000))
    x += 1
    print(f'{x} / 2306')
    #this part takes about 30 minutes to complete
    
    
    



```

```python
#Nucleotide Sequences for All Genes
protein_nt_seq = []
y=0
for i in TK_GENES:  
    URL = "https://www.genome.jp/entry/-f+-n+n+tko:" + i
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    AA = soup.find("pre")
    TKAA=str(AA)
    Almost = TKAA.replace('</pre>', '')
    Almost_Info = (Almost.rsplit('(N)')[1])
    protein_nt_seq.append(Almost_Info.replace("\n", ""))
    y+=1
    print(y)
```

```python
#Put in Table and filter protein sizes
sequence_strings = [''.join(ele) for ele in protein_sequence]

Tko_Protein_MW["Description"] = protein_info
Tko_Protein_MW["MW (kDa)"] = protein_value
Tko_Protein_MW["AA_Sequence"] = sequence_strings
Tko_Protein_MW["NT_Sequence"] = protein_nt_seq



Tko_Protein_MW.to_csv('Tko_Protein_Information.csv', index=False)
```
