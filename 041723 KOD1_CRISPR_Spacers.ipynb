```python
#import packages
import numpy as np
import pandas as pd
import re
import Bio

#I saved the TKO genome word file as a txt file with just the sequence
#import Tko sequence and as a string
KOD1 = open("KOD1 Thermococcus Genome.txt")
KOD1 = KOD1.read()
print(len (KOD1) )

#import Cas spacer sequences, find out where they are located in the genome
CRISPIR_locus_1 = "TTTCAATTCTCCTAGAGTCTTATTGCAAC" #15 Results
print(re.search("TTTCAATTCTCTTAGAGTCTTATTGCAAC(.*)TTTCAATTCTCTTAGAGTCTTATTGCAAC" , KOD1))

CRISPR_locus_2 = "GTTTCAATTCTCTTAGAGTCTTATTGCAAC" #19 Results
print(re.search("GTTTCAATTCTCTTAGAGTCTTATTGCAAC(.*)GTTTCAATTCTCTTAGAGTCTTATTGCAAC" , KOD1))

CRISPR_locus_3 = "GTTGCAATAAGACTCTAAGAGAATTGAAAG" #27 Results
print(re.search("GTTGCAATAAGACTCTAAGAGAATTGAAAG(.*)GTTGCAATAAGACTCTAAGAGAATTGAAAG" , KOD1))


#make list of invader sequences for each locus
KOD1_Cas1 = KOD1[373033 : 374071]
KOD1_Cas1_split = KOD1_Cas1.split(CRISPIR_locus_1)
#print(KOD1_Cas1_split)
#print(len(KOD1_Cas1_split))

KOD1_Cas2 = KOD1[469085 : 470603]
KOD1_Cas2_split = KOD1_Cas2.split(CRISPR_locus_2)
#print(KOD1_Cas2_split)
#print(len(KOD1_Cas2_split))

KOD1_Cas3 = KOD1[833560 : 835982]
KOD1_Cas3_split = KOD1_Cas3.split(CRISPR_locus_3)
#print(KOD1_Cas3_split)
#print(len(KOD1_Cas3_split))

#convert each list into a dataframe and consolidate
KOD1_Spacers_df1=pd.DataFrame(KOD1_Cas1_split)
KOD1_Spacers_df1.rename(columns={0: "Spacer Sequence"}, inplace=True)
Locus1 = []
for _ in range (0, len(KOD1_Spacers_df1)):
    _ = "CRISPR_1"
    Locus1.append(_)
KOD1_Spacers_df1["Locus"]=Locus1

KOD1_Spacers_df2=pd.DataFrame(KOD1_Cas2_split)
Locus2 = []
KOD1_Spacers_df2=pd.DataFrame(KOD1_Cas2_split)
for i in range (0, len(KOD1_Spacers_df2)):
    i = "CRISPR_2"
    Locus2.append(i)
KOD1_Spacers_df2["Locus"]=Locus2
KOD1_Spacers_df2.rename(columns={0: "Spacer Sequence"}, inplace=True)

KOD1_Spacers_df3=pd.DataFrame(KOD1_Cas3_split)
Locus3 = []
KOD1_Spacers_df3=pd.DataFrame(KOD1_Cas3_split)
for z in range (0, len(KOD1_Spacers_df3)):
    z = "CRISPR_3"
    Locus3.append(z)
KOD1_Spacers_df3["Locus"]=Locus3
KOD1_Spacers_df3.rename(columns={0: "Spacer Sequence"}, inplace=True)

data= [KOD1_Spacers_df1, KOD1_Spacers_df2, KOD1_Spacers_df3]
KOD1_Spacers = pd.concat(data, ignore_index=True)


print(KOD1_Spacers)

KOD1_Spacers.to_csv("KOD1_Spacers.csv")
```
